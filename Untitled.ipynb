Let us first redefine the methods we used for defining the RRT in tutorial 4

```python
#visibility prm

def coll(q):
     '''Return true if q in collision, false otherwise.'''
     pin.updateGeometryPlacements(robot.model,robot.data,robot.collision_model,robot.collision_data,q)
     return pin.computeCollisions(robot.collision_model,robot.collision_data,False)

def RAND_CONF(checkcollision=True):
    '''
    Return a random configuration, not is collision
    '''
    while True:
        q = pin.randomConfiguration(robot.model)  # sample between -3.2 and +3.2.
        if not (checkcollision and coll(q)): return q

# %load tp4/generated/rrt_nearest
def distance(q1,q2):    
    '''Return the euclidian distance between two configurations'''
    return np.linalg.norm(q2-q1)
        
def NEAREST_VERTEX(G,q_rand):
    '''returns the index of the Node of G with the configuration closest to q_rand  '''
    min_dist = 10e4
    idx=-1
    for (i,node) in enumerate(G):
        dist = distance(node[1],q_rand) 
        if dist < min_dist:
            min_dist = dist
            idx = i
    return idx

def ADD_EDGE_AND_VERTEX(G,parent,q):
    G += [(parent,q)]

def lerp(q0,q1,t):    
    return q0 * (1 - t) + q1 * t

def NEW_CONF(q_near,q_rand,discretisationsteps, delta_q = None):
    '''Return the closest configuration q_new such that the path q_near => q_new is the longest
    along the linear interpolation (q_near,q_rand) that is collision free and of length <  delta_q'''
    q_end = q_rand.copy()
    dist = distance(q_near, q_rand)
    if delta_q is not None and dist > delta_q:
        #compute the configuration that corresponds to a path of length delta_q
        q_end = lerp(q_near,q_rand,delta_q/dist)
        dist = delta_q
    dt = dist / discretisationsteps
    for i in range(1,discretisationsteps):
        q = lerp(q_near,q_end,dt*i)
        if coll(q):
            return lerp(q_near,q_end,dt*(i-1))
    return q_end


def VALID_EDGE(q_new,q_goal,discretisationsteps=400):
    return np.linalg.norm(q_goal -NEW_CONF(q_new, q_goal,discretisationsteps)) < 1e-3

discretisationsteps_newconf = 200 #To tweak later on
discretisationsteps_validedge = 200 #To tweak later on
k = 1000  #To tweak later on
delta_q = 3. #To tweak later on


q_init = robot.q0.copy()
```

Now to define the visibility PRM

```python
M = 200
ntry = 0
guards_CC = []
connection = []
guards = []
def merge(guards_CC, idx1, idx2, q_rand, g, g_vis ):
    G = guards_CC[idx1]
    ADD_EDGE_AND_VERTEX(G,len(G)-1,q_rand)
    for egde_vert in guards_CC[idx2]:
        ADD_EDGE_AND_VERTEX(G,len(G)-1,egde_vert[1])
    guards_CC[idx1] = G
    guards_CC.pop(idx2)
    return guards_CC

while ntry < M:
    q_rand = RAND_CONF()
    g_vis = None
    G_vis = None
    G_vis_idx = None
    for i, G_i in enumerate(guards_CC):
        for (_, g) in G_i:
            if VALID_EDGE(g, q_rand):
                if g_vis is None:
                    g_vis = g
                    G_vis = G_i
                    G_vis_idx = i
                else:
                    #merge(G_vis, G_i,G_vis_idx, i, q_rand, g, g_vis)
                    guards_CC = merge(guards_CC,G_vis_idx, i, q_rand, g, g_vis)
                break
    if g_vis is None:
        guards_CC += [[(None,q_rand)]]
        guards += [q_rand]
        ntry = 0
    else:
        ntry+=1
    
```

```python
import matplotlib.pyplot as plt

def plot_graph(guards_cc, guards, data = distance_data, markerSize=20):
    """
    Plots points and draws edges between them based on the parentId in the graph.
    
    Parameters:
    graph (list): List of tuples where each tuple is (parentId, point).
                  - parentId is an integer pointing to the index of the parent point or None if no parent.
                  - point is a tuple representing the (x, y) coordinates of the point.
    """
    plt.scatter(data[:,0],data[:,1],c=data[:,2],s=markerSize,lw=0)
    for i, graph in enumerate(guards_cc):
        # Plot points
        for j, (parentId, point) in enumerate(graph):
            x, y = point
            plt.scatter(x, y, color='blue')
    
            # Draw an edge if the point has a parent
            if parentId is not None:
                parent_point = graph[parentId][1]
                px, py = parent_point
                plt.plot([x, px], [y, py], 'k-', lw=1)  # Draw edge between the point and its parent
    
    # Display the plot
    for guard in guards:
        plt.scatter(guard[0], guard[1], color='red')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Graph Plot - guards in red")
    plt.show()
```
